```python
import pandas as pd
```

```python
df_500 = pd.read_csv('imdb_movies_detail_500.csv')
df_1000 = pd.read_csv('imdb_movies_detail_1000.csv')
df_2000 = pd.read_csv('imdb_movies_detail_2000.csv')
df_2416 = pd.read_csv('imdb_movies_detail_2416.csv')
```

```python
df = pd.concat([df_500, df_1000, df_2000, df_2416], ignore_index=True)
```

```python
df.shape
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df_id = pd.read_csv('imdb_movies.csv')
```

```python
df_id.head()
```

```python
# merge df and df_id with id and imdb_id
df_total = pd.merge(df, df_id, left_on='imdb_id', right_on='id', how='inner')
```

```python
df_total.head()
```

```python
df_total.shape
```

```python
df_total.to_csv('imdb_movies_detail_complete.csv', index=False)
```
