```python
import pandas as pd
```

```python
df_500 = pd.read_csv('../data/raw/imdb_movies_detail_500.csv')
df_1000 = pd.read_csv('../data/raw/imdb_movies_detail_1000.csv')
df_2000 = pd.read_csv('../data/raw/imdb_movies_detail_2000.csv')
df_2416 = pd.read_csv('../data/raw/imdb_movies_detail_2416.csv')
```

```python
df = pd.concat([df_500, df_1000, df_2000, df_2416], ignore_index=True)
```

```python
df.shape
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df_id = pd.read_csv('../data/raw/imdb_movies.csv')
df_id.head()
```

```python
df_yr = df_id[df_id['year'] != df_id['runtime']]
df_yr.shape
```

```python
# drop rating and cotes from df_id
df_id.drop(['rating', 'votes', 'plot', 'runtime'], axis=1, inplace=True)
df_id.head()
```

```python
print(df.columns)
print(df.shape)
print(df_id.columns)
print(df_id.shape)
```

```python
# merge df and df_id with id and imdb_id
df_complete = pd.merge(df_id, df, left_on='id', right_on='imdb_id', how='left')
print(df_complete.shape)
print(df_complete.columns)
```

```python
df_complete.to_csv('../data/raw/imdb_movies_complete.csv', index=False)
```

```python
df_missing = pd.read_csv('../data/raw/imdb_movies_detail_500_new.csv')
df_missing.head()
```

```python
df_missing['summary'].isnull().sum()
```

```python
df_missing['synopsis'].isnull().sum()
```

```python
print(df_complete.columns)
print(df_missing.columns)
```

```python
df_complete.columns
```

```python
# create new df with only imdb_id, summary and synopsis from df_missing
df_missing = df_missing[['imdb_id', 'summary', 'synopsis']]
```

```python
print(df_complete['synopsis'].isnull().sum())
print(df_complete['summary'].isnull().sum())
```

```python
# Merge the dataframes
df_complete = pd.merge(df_complete, df_missing, on='imdb_id', how='left', suffixes=('', '_missing'))

# Update 'summary' and 'synopsis' columns
df_complete['summary'] = df_complete['summary_missing'].combine_first(df_complete['summary'])
df_complete['synopsis'] = df_complete['synopsis_missing'].combine_first(df_complete['synopsis'])

# Drop the extra columns
df_complete = df_complete.drop(columns=['summary_missing', 'synopsis_missing'])
```

```python
print(df_complete['synopsis'].isnull().sum())
print(df_complete['summary'].isnull().sum())
```

```python
df_complete.shape
```

```python
df_complete.columns
```

```python
df_complete.to_csv('../data/raw/imdb_movies_complete.csv', index=False)
```
